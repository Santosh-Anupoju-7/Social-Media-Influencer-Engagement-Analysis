```python
#Checking Duplicate values
df.duplicated().sum()
```

```python
# Checking Outliers

for i in df.columns:
  if (df[i].dtype != 'object') and (df[i].dtype != 'bool'):
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
df.columns
```

```python
#Handeling Outliers

outlier_col = ['followers', 'following', 'likes', 'videos_count', 'awg_engagement_rate', 'comment_engagement_rate', 'like_engagement_rate']

for i in outlier_col:
  q1 = df[i].quantile(0.25)
  q3 = df[i].quantile(0.75)
  IQR = q3-q1
  Lower_limit = q1 - 1.5 * IQR
  upper_limit = q3 +  1.5 * IQR
  df = df[(df[i]>=Lower_limit) & (df[i]<=upper_limit)]
```

```python
#After Handled the Outliers
for i in df.columns:
  if (df[i].dtype != 'object') and (df[i].dtype != 'bool'):
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

**VBA: Visual Basic for Applications.**

```python
#Top 10 Influencers by Followers

top = df.sort_values(by='followers', ascending=False).head(10)

sns.barplot(data=top, x='followers', y='nickname')
plt.title("Top 10 Influencers by Followers")
plt.xlabel("Followers")
plt.ylabel("Influencer")
plt.show()
```

```python
#Followers vs Engagement

sns.scatterplot(data=df, x='followers', y='awg_engagement_rate')
plt.title("Followers vs Average Engagement Rate")
plt.xlabel("Followers")
plt.ylabel("Average Engagement Rate")
plt.show()
```

```python
#Verfied vs Non-verfied Engagement
sns.boxplot(data=df, x='is_verified', y='awg_engagement_rate')
plt.title("Verified vs Non-Verified Engagement")
plt.xlabel("Verified")
plt.ylabel("Average Engagement Rate")
plt.show()
```

```python
# Distribution of Followers
sns.histplot(data=df, x='followers')
plt.title("Distribution of Followers")
plt.xlabel("Followers")
plt.ylabel("Frequency")
plt.show()
```

```python
#Likes vs Videos Count

sns.scatterplot(data=df, x='likes', y='videos_count')
plt.title("Likes vs Videos Count")
plt.xlabel("Likes")
plt.ylabel("Videos Count")
plt.show()
```

```python
#correlation

corr = df[['followers','following','likes','videos_count','awg_engagement_rate','like_engagement_rate']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
